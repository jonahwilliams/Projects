<h1>Urban/Suburban Poverty Analysis using ACS data</h1>
<p>BGD on a logit model - predict percent of population in poverty given other percentage variables for census block groups in Iowa and Nebraska.  This is an exploratory project that aims to identify links between poverty in urban and suburban enviornments, wtih respect to their differences and neighborhoods which are difficult to model</p>

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
%matplotlib inline
%load_ext rmagic 

df = pd.read_excel('/Users/jonahwilliams/Downloads/census blck_2014.xlsx')
df.index = df.GIDBG
    
for col in df.columns:
    if "MOE" in col:
        df = df.drop([col], axis=1)
    elif "CEN" in col:
        df = df.drop([col], axis=1)
    elif "pct" not in col and "area code" not in col:
        df = df.drop([col], axis=1)
    else:
        pass
        
df = df[(df['area code']==1) | (df['area code']==2)] \
       .drop(['pct_Females_ACS_08_12',
              'pct_Pop_65plus_ACS_08_12',
              'pct_Pov_Univ_ACS_08_12',
              'pct_Pop_5yrs_Over_ACS_08_12',
              'pct_Rel_Under_6_ACS_08_12',
              'pct_Pop_1yr_Over_ACS_08_12',
              'pct_Not_MrdCple_HHD_ACS_08_12',
              'pct_Owner_Occp_HU_ACS_08_12'],
               axis=1)

    
df['area code'] = df['area code'].map(lambda x: int(x == 1))
df.columns = ["".join(col.split("_")[1:-3]) if col != "area code" else col for col in df.columns]
%Rpush df
df.head()
```

```python
%%R -w 50 -h 25 -u cm
library(ggplot2)
library(GGally)
df$area.code <- factor(df$area.code)
ggparcoord(df, columns = 2:44, groupColumn=1, alpha=1) + theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

<h3>Variation between Urban and Suburban Areas</h3>
<p>Above we have several parallel coordinate plots which allow us the see the variation among urban and suburban regions with respect to each variable present.  Note that there will be several variables for which there exists higher variation and/or a higher mean in one group and not in the other.  Mobile Homes for instance have a higher variance in suburban areas(i.e. there are actually suburban mobile homes but no urban ones).

```python
std_df = df
for col in std_df.columns:
    if col != "area code" and col != "PrsBlwPovLev":
        std_df[col] = (std_df[col] - np.mean(std_df[col]))/ (np.std(std_df[col]))
%Rpush std_df
```

```python
%%R -w 50 -h 50 -u cm
library(reshape2)
data <- melt(std_df, id.vars=c("area.code","PrsBlwPovLev"))
ggplot(data=data,aes(x=value, y=PrsBlwPovLev, color=factor(area.code), group=variable)) + 
    geom_point() + facet_wrap(~variable, scales="free")
```

<h3>Scatter Matrix of standardized covariates versus standardized poverty, colored by urban versus suburban</h3>
<p>The most obvious (linear) relationship that holds above is an increase in the number of married households in urban and suburban correlates nicely with a decrease in poverty.  There are many more examples where there is high variability in once group, such as RecentBuiltHU in Suburban and little to no variability in Urban.  This exposes problems for a model with a simple additive term, and implies the need for multiplicative interactions.

$ log(\frac{y}{1 - y}) = \beta_0 + \sum \limits_{j=0}^p \beta_{j}\frac{(X_j - \bar{X_j})}{\sigma(X_{j})}$

First we will fit a simple additive model, then move towards one with interaction terms if the fit is poor enough

```python
std_df["PrsBlwPovLev"] = std_df["PrsBlwPovLev"] + 0.001
std_df["PrsBlwPovLev"] = np.log(std_df["PrsBlwPovLev"]/ (100 - std_df["PrsBlwPovLev"]))
std_df = std_df.dropna(axis=0)
X = std_df[[col for col in std_df.columns if col != "PrsBlwPovLev"]].values
y = std_df[["PrsBlwPovLev"]].values
```

```python
clf = linear_model.Lasso(alpha=0.04)
clf.fit(X, y)
est_y = clf.predict(X)
plt.figure(figsize=(12, 8))
plt.title("Distribution of Residuals")
plt.hist(y[0::,0] - est_y, histtype="stepfilled", bins=100)
plt.show()
```

<h3>Strong bimodal structure in residuals</h3>
<p>While the majority of the error residuals seem to be distributed about 0 approximately symmetrically, there is a subgroup with much higher mean error.  The sign of the error inplies severe overestimation from our model</p>

```python
parameters = []
errors = []
for i in np.linspace(0.01, 1, 100):
    error = []
    for j in range(len(X)):
        mask = np.array([True if k != j else False for k in range(len(X))])
        clf = linear_model.Lasso(alpha=i)
        clf.fit(X[np.ix_(mask)], y[np.ix_(mask)])
        error.append(np.square(y[j, 0] - clf.predict(X[j, 0::])))
    errors.append(np.mean(error))
    clf = linear_model.Lasso(alpha=i)
    clf.fit(X, y)
    parameters.append(clf.coef_)
```

```python
parameters = np.array(parameters)
colors = sns.color_palette("hls", 12)
fig, ax = plt.subplots(4, 1, sharex=True, sharey=False)

fig.suptitle("Coefficients versus Regularization Strength", fontsize=20)
fig.set_size_inches(18.5, 12)
for i in range(0, 12):
    ax[0].plot(np.linspace(0.01, 1, 100), parameters[0::, i],
               label=std_df.columns[i], color=colors[i - 0])
ax[0].legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

for i in range(12, 24):
    ax[1].plot(np.linspace(0.01, 1, 100), parameters[0::, i],
               label=std_df.columns[i], color=colors[i - 12])
ax[1].legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

for i in range(24, 36):
    ax[2].plot(np.linspace(0.01, 1, 100), parameters[0::, i],
               label=std_df.columns[i], color=colors[i - 24])
ax[2].legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

for i in range(36, 43):
    ax[3].plot(np.linspace(0.01, 1, 100), parameters[0::, i],
               label=std_df.columns[i], color=colors[i - 36])
ax[3].legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

<h3>Selection of important parameters</h3>
<p>With just a small amount of regularization, almost all the parameters from the first graph go towards zero.  We can calculate the optimal regularization hyper-parameter $\lambda$ by minimizing the cross validated error.  Here I choose the exclude $\lambda = 0$</p>

```python
plt.figure(figsize=(12, 8))
plt.plot(np.linspace(0.01, 1, 100), errors)
l = np.argmin(errors[1::])
plt.scatter(np.linspace(0.01, 1, 100)[l + 1], errors[l + 1],
            color="red",
            label="minimum error: "+ str(i + 1) + ": "+ str(np.linspace(0.01, 1, 100)[l + 1]))
plt.legend()
plt.xlabel("$\lambda$")
plt.ylabel("LOO MSE")
plt.xlim(0, 1)
plt.show()
```

<p>While there is a general increasing trend in the MSE, there is a small decline with some moderate regularization strength.  This value will be chosen to select out some of the parameters</p>

```python
clf = linear_model.Lasso(alpha=0.04)
clf.fit(X, y)
positions = [i for i in range(len(clf.coef_))]
values  = [c for c in clf.coef_]
labels = [std_df.columns[i] for i in range(len(clf.coef_))]

fig, ax = plt.subplots()

fig.set_size_inches(18.5, 6)
for i in range(len(clf.coef_)):
    if np.abs(clf.coef_[i]) >  0.01:
        rect = ax.bar(positions[i], values[i], label=labels[i], color="grey")[0]
        height = rect.get_height()
        if clf.coef_[i] < 0:
            plt.text(rect.get_x()+rect.get_width()/2., -1*height - 0.05, '%s'%str(clf.coef_[i])[0:4],
                ha='center', va='bottom')
        else:
            plt.text(rect.get_x()+rect.get_width()/2., height + 0.01, '%s'%str(clf.coef_[i])[0:4],
                ha='center', va='bottom')
    else:
        ax.bar(positions[i], values[i], label=labels[i], color="grey", alpha=0.5)
plt.xticks(positions, labels, rotation='vertical')
plt.title("Parameter Estimates")
plt.show()
```

<h3>Bimodal Structure Emerges in Urban Residuals</h3>

<p>Here we have a distribution of the residuals from the two models, where overprediction of poverty results in a negative value and underprediction results in a positive value.  Note the clear bimodal struture in the urban residuals.  There is a subgroup for which our model is not correct, in the sense that it consistently overpredicts the poverty percentages.</p>
