```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
%matplotlib inline
%load_ext rmagic 

df = pd.read_excel('/Users/jonahwilliams/Projects/ACS/census_data.xlsx')
df.index = df.GIDBG
    
for col in df.columns:
    if "MOE" in col:
        df = df.drop([col], axis=1)
    elif "CEN" in col:
        df = df.drop([col], axis=1)
    elif "pct" not in col and "area code" not in col:
        df = df.drop([col], axis=1)
    else:
        pass
        
df = df[(df['area code']==1) | (df['area code']==2)] \
       .drop(['pct_Females_ACS_08_12',
              'pct_Pop_65plus_ACS_08_12',
              'pct_Pov_Univ_ACS_08_12',
              'pct_Pop_5yrs_Over_ACS_08_12',
              'pct_Pop_1yr_Over_ACS_08_12'],
               axis=1)

    
df['area code'] = df['area code'].map(lambda x: int(x == 1))
df.columns = ["".join(col.split("_")[1:-3]) if col != "area code" else col for col in df.columns]
%Rpush df
df.head()
```

<h3>Scatter Matrix of standardized covariates versus standardized poverty, colored by urban versus suburban</h3>
<p>The most obvious (linear) relationship that holds above is an increase in the number of married households in urban and suburban correlates nicely with a decrease in poverty.  There are many more examples where there is high variability in once group, such as RecentBuiltHU in Suburban and little to no variability in Urban.  This exposes problems for a model with a simple additive term, and implies the need for multiplicative interactions.

$ log(\frac{y}{1 - y}) = \beta_0 + \sum \limits_{j=0}^p \beta_{j}\frac{(X_j - \bar{X_j})}{\sigma(X_{j})}$

First we will fit a simple additive model, then move towards one with interaction terms if the fit is poor enough

```python
df["PrsBlwPovLev"] = df["PrsBlwPovLev"] + 0.001
df["PrsBlwPovLev"] = np.log(df["PrsBlwPovLev"]/ (100 - df["PrsBlwPovLev"]))
df = df.dropna(axis=0)
```

```python
X = df[[col for col in df.columns if col != "PrsBlwPovLev" and col != "area code"]].values
y = df[["PrsBlwPovLev"]].values
clf = linear_model.Lasso(alpha=0.01)
clf.fit(X, y)
est_y_1 = clf.predict(X)
plt.figure(figsize=(12, 8))
plt.title("Distribution of Residuals")
plt.hist(y[0::,0] - est_y_1, histtype="stepfilled", bins=100)
plt.show()
```

```python
for col in df.columns:
    if col != "area code" and col != "PrsBlwPovLev":
        df["Urban" + str(col)] = df["area code"] * df[col]

df.head()
```

```python
X = df[[col for col in df.columns if col != "PrsBlwPovLev" and col != "area code"]].values
y = df[["PrsBlwPovLev"]].values
clf = linear_model.Lasso(alpha=0.01)
clf.fit(X, y)
est_y = clf.predict(X)
plt.figure(figsize=(12, 8))
plt.title("Distribution of Residuals")
plt.hist(y[0::,0] - est_y, histtype="stepfilled", bins=200, alpha=0.3, color="blue")
plt.hist(y[0::,0] - est_y_1, histtype="stepfilled", bins=200, alpha=0.3, color="red")
plt.show()
```

```python
sns.set_style("darkgrid")

positions = [i for i in range(len(clf.coef_))]
values  = [c for c in clf.coef_]
labels = [df.columns[i] for i in range(len(df.columns)) if df.columns[i] != "PrsBlwPovLev"]

positions = [i for i in range(42)]
values = [clf.coef_[i] for i in range(42)]
urban_values = [clf.coef_[i] + clf.coef_[i + 41]  for i in range(42)]

labels = [df.columns[i] for i in range(44) if df.columns[i] != "PrsBlwPovLev" and df.columns[i] != "area code"]


fig, ax = plt.subplots()
fig.set_size_inches(8, 12)

ax.barh(np.array(positions), values, color="#377eb8", height=0.45)
ax.barh(np.array(positions) + 0.35, urban_values, color="#e41a1c", height=0.45)


patch_1 = mpatches.Patch(color="#377eb8", label='Suburban')
patch_2 = mpatches.Patch(color="#e41a1c", label='Urban')
plt.legend(handles=[patch_1, patch_2])

plt.title("Parameter Estimates")
plt.yticks(np.array(positions), labels)
plt.show()
```

```python
from json import loads
from descartes import PolygonPatch
import matplotlib.patches as mpatches
import matplotlib as mpl
import matplotlib.cm as cm

json_data = loads(open("nebraska.json","r").read())['features']
json_data2 = loads(open("iowa.json","r").read())['features']

json_data = json_data + json_data2
for item in json_data:
    lookup = item["properties"]["GEOID"]
    y = df[df.index == int(lookup)]["PrsBlwPovLev"].values
    X = df[df.index == int(lookup)][[col for col in df.columns if col != "PrsBlwPovLev" and col != "area code"]].values
    r = y - clf.predict(X) 
    item["properties"]["residual"] = r   
    

cmaps = ["#ca0020",
"#f4a582",
"#f7f7f7",
"#92c5de",
"#0571b0",
"#545454"]

def residualMap(x):
    if (x >= -1.5) and (x <= 1.5):
        return 2
    elif (x > 1.5) and (x <= 3):
        return 3
    elif (x > 3):
        return 4
    elif (x < -1.5) and (x >= -3):
        return 1
    elif (x < -3):
        return 0
    else:
        return 5


fig = plt.figure(figsize=(100, 100)) 
ax = fig.gca()
sns.set_style("white")

for h in range(len(json_data)):
    poly = {'type': 'Polygon', 'coordinates': json_data[h]["geometry"]["coordinates"]}
    fill = residualMap(json_data[h]['properties']['residual'])
    ax.add_patch(PolygonPatch(poly,ec="black",fc=cmaps[fill], alpha=1, zorder=2 ))

ax.axis('scaled')
```
