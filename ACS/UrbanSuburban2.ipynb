<h1>Urban/Suburban Poverty Analysis using ACS data</h1>
<p>BGD on a logit model - predict percent of population in poverty given other percentage variables for census block groups in Iowa and Nebraska.  This is an exporatory project that aims to identify links between poverty in urban and suburban enviornments, wtih respect to their differences and neighborhoods which are difficult to model</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
%matplotlib inline
%load_ext rmagic 

df = pd.read_excel('/Users/jonahwilliams/Downloads/census blck_2014.xlsx').fillna(0)

def removeInf(x):
    if x == float('inf'):
        x = 0
    return x

urban = df[[col for col in df.columns if 'pct' in col and 'MOE' not in col and "CEN" not in col]][df["area code"]==1]
urban = urban.applymap(lambda x: float(x)/100).reset_index().drop('index',axis=1)
urban = urban.drop(['pct_Females_ACS_08_12',
                              'pct_Pop_65plus_ACS_08_12',
                              'pct_Pov_Univ_ACS_08_12',
                              'pct_Pop_5yrs_Over_ACS_08_12',
                              'pct_Rel_Under_6_ACS_08_12',
                              'pct_Pop_1yr_Over_ACS_08_12',
                              'pct_Not_MrdCple_HHD_ACS_08_12'],
                               axis=1)

suburban = df[[col for col in df.columns if 'pct' in col and 'MOE' not in col and "CEN" not in col]][df["area code"]==2]
suburban = suburban.applymap(lambda x: float(x)/100).reset_index().drop('index',axis=1)
suburban = suburban.drop(['pct_Females_ACS_08_12',
                              'pct_Pop_65plus_ACS_08_12',
                              'pct_Pov_Univ_ACS_08_12',
                              'pct_Pop_5yrs_Over_ACS_08_12',
                              'pct_Rel_Under_6_ACS_08_12',
                              'pct_Pop_1yr_Over_ACS_08_12',
                              'pct_Not_MrdCple_HHD_ACS_08_12'],
                               axis=1)
param_lookup = {}
param_mapper = {}
for i, col in enumerate(suburban.columns):
    param_lookup[i] = " ".join(col.split("_")[0:-3])
    param_mapper[col] = i
```

<h3>Scatter Matrix of standardized covariates versus standardized poverty, colored by urban versus suburban</h3>
<p>The most obvious (linear) relationship that holds above is an increase in the number of married households in urban and suburban correlates nicely with a decrease in poverty.  There are many more examples where there is high variability in once group, such as RecentBuiltHU in Suburban and little to no variability in Urban.  This exposes problems for a model with a simple additive term, and implies the need for multiplicative interactions.

$ E[Y] = g(\beta_0 + \sum \limits_{j=0}^p \beta_{j,1}X_j + \sum \limits_{j=0}^p \beta_{j,2} X_j X_{urban})$

First we will fit a simple additive model, then move towards one with interaction terms if the fit is poor enough

```python
filtered["urban"] = filtered["urban"].map({"Suburban":0,"Urban":1})
filtered["PrsBlwPovLevACS"] = np.log(filtered["PrsBlwPovLevACS"])
X = filtered[[col for col in filtered.columns if col != "PrsBlwPovLevACS"]].values
y = filtered[["PrsBlwPovLevACS"]].values
```

```python
def logit(data, parameters, bias):
    return 1 / (1 + np.exp(-(np.dot(data, parameters) + bias)))

sub_parameters = []
sub_biases = []
sub_errors = []
for i in range(100):
    error = []
    param = np.random.normal(0, 0.01, (len(X[0])))
    bias = 0
    for i in xrange(1000):
        if (i % 100 == 0):
            error.append(np.mean((logit(X, param, bias) - y[0::, 0])**2))
        param += -0.01 * np.dot((logit(X, param, bias) - y[0::, 0]) * (1 - logit(X, param, bias) * logit(X, param, bias)), X)
        bias += -0.01 * np.mean((logit(X, param, bias) - y[0::, 0]) * (1 - logit(X, param, bias) * logit(X, param, bias)))
    sub_errors.append(error)
    sub_parameters.append(param)
    sub_biases.append(bias)
```

```python
fig, ax = plt.subplots(1, 1, sharex=False, sharey=True)
fig.set_size_inches(20,10.5)
fig.suptitle("Training Error", fontsize=20)


for item in sub_errors:
    ax.plot(item, color="grey", alpha=0.2)
ax.set_xlabel("100th Training Iteration")
plt.show()
```

```python
urb_param = dict((i,[]) for i in range(len(X[0])))
sub_param = dict((i,[]) for i in range(len(X[0])))
mean_urb_param = dict((i,[]) for i in range(len(X[0])))
mean_sub_param = dict((i,[]) for i in range(len(X[0])))
for i in range(100):
    for j in range(len(X[0])):
        sub_param[j].append(sub_parameters[i][j])
        urb_param[j].append(urb_parameters[i][j])

for key, value in urb_param.items():
    mean_urb_param[key] = np.mean(value)

for key, value in sub_param.items():
    mean_sub_param[key] = np.mean(value)
```

<h1>Residual Analysis</h1>
Because the model was chosen and fit on all of the data, we can not get an unbiased estimate of out of sample error.  This was outside the scope of the project however, especially due to the small amount of data.  Instead we will use the optimal parameter estimates and look for local co-occurance of high residual values, essentailly identifying neighborhoods with undocumented but significant local explanatory variables.

```python
def calc_mode(x):
    '''Approximates the mode of the data by resampling and using numpy.mode'''
    from scipy.stats.mstats import mode
    return mode(np.random.choice(x, 100000))[0]

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(12, 8)
urban_model = np.zeros(len(X[0]))
for i in range(len(X[0])):
    urban_model[i] = calc_mode(np.array(urb_parameters)[0::, i])
urban_bias = calc_mode(urb_biases)
y = urban[["pct_Prs_Blw_Pov_Lev_ACS_08_12"]].values
X = urban[[col for col in urban.columns if col != "pct_Prs_Blw_Pov_Lev_ACS_08_12"]].values

ax[0].hist(y[0::, 0] - logit(X, urban_model, urban_bias), histtype="stepfilled", bins=100)
ax[0].set_xlabel("Urban residuals")


suburban_model = np.zeros(len(X[0]))
for i in range(len(X[0])):
    suburban_model[i] = np.mean(np.array(sub_parameters)[0::, i])
suburban_bias = np.mean(sub_biases)
y = suburban[["pct_Prs_Blw_Pov_Lev_ACS_08_12"]].values
X = suburban[[col for col in suburban.columns if col != "pct_Prs_Blw_Pov_Lev_ACS_08_12"]].values

ax[1].hist(y[0::, 0] - logit(X, suburban_model, suburban_bias), histtype="stepfilled", bins=100)
ax[1].set_xlabel("Suburban residuals")
fig.suptitle('Error Distributions for Urban and Suburban Models', fontsize=20)
plt.show()
```

```python
from json import loads
from descartes import PolygonPatch

json_data = loads(open("censusBlockData.json","r").read())['features']
df.index = df.GIDBG
lookup = df[[col for col in df.columns if 'pct' in col and 'MOE' not in col and "CEN" not in col]]\
        .applymap(lambda x: float(x)/100)\
        .drop(['pct_Females_ACS_08_12',
              'pct_Pop_65plus_ACS_08_12',
              'pct_Pov_Univ_ACS_08_12',
              'pct_Pop_5yrs_Over_ACS_08_12',
              'pct_Rel_Under_6_ACS_08_12',
              'pct_Pop_1yr_Over_ACS_08_12'],
               axis=1)

for h in range(len(json_data)):
    if json_data[h]['properties']['area code'] == 1:
        filtered = lookup[lookup.index == int(json_data[h]['properties']['GEOID'])]
        res = filtered[["pct_Prs_Blw_Pov_Lev_ACS_08_12"]].values[0] - logit(filtered[[col for col in filtered.columns if col != "pct_Prs_Blw_Pov_Lev_ACS_08_12"]].values[0],
              urban_model, urban_bias)
        json_data[h]['properties']['residual'] = res
    if json_data[h]['properties']['area code'] == 2:
        filtered = lookup[lookup.index == int(json_data[h]['properties']['GEOID'])]
        res = filtered[["pct_Prs_Blw_Pov_Lev_ACS_08_12"]].values[0] - logit(filtered[[col for col in filtered.columns if col != "pct_Prs_Blw_Pov_Lev_ACS_08_12"]].values[0],
              suburban_model, suburban_bias)
        json_data[h]['properties']['residual'] = res
        
        
```

```python
# Source : Colorbrewer2.org
colormap = ["#b2182b", "#ef8a62", "#fddbc7", "#f7f7f7", "#d1e5f0", "#67a9cf", "#2166ac"]

urban_std = np.std(urban_residual)
suburban_std = np.std(suburban_residual)

def closest_breakpoint(x, std):
    if (x > 0 - std) and (x < 0 + std):
        return 3
    elif (x >= 0 + std) and (x < 0 + 2*std):
        return 4
    elif (x >= 0 + 2*std) and (x < 0 + 3*std):
        return 5
    elif (x >= 0 + 3*std):
        return 6
    elif (x <= 0 - std) and (x > 0 - 2*std):
        return 2
    elif (x <= 0 - 2*std)and (x > 0 - 3*std):
        return 1
    else:
        return 0
```

```python
sns.set_style("white")
import matplotlib.patches as mpatches
fig = plt.figure(figsize=(20, 20)) 
ax = fig.gca() 
for h in range(len(json_data)):
    if json_data[h]['properties']['area code'] == 1:
        poly = {'type': 'Polygon', 'coordinates': json_data[h]["geometry"]["coordinates"]}
        fill = closest_breakpoint(json_data[h]['properties']['residual'], urban_std)
        ax.add_patch(PolygonPatch(poly,ec="black",fc=colormap[fill], alpha=1, zorder=2 ))
    if json_data[h]['properties']['area code'] == 2:
        fill = closest_breakpoint(json_data[h]['properties']['residual'], suburban_std)
        poly = {'type': 'Polygon', 'coordinates': json_data[h]["geometry"]["coordinates"]}
        ax.add_patch(PolygonPatch(poly,ec="black",fc=colormap[fill],alpha=1, zorder=2 ))
ax.set_xlim(-97, -95)
ax.set_ylim(40.75, 41.9)


patch1 = mpatches.Patch(color=colormap[0], label='$-3\sigma$')
patch2 = mpatches.Patch(color=colormap[1], label='$-2\sigma$')
patch3 = mpatches.Patch(color=colormap[2], label='$-\sigma$')
patch4 = mpatches.Patch(color=colormap[3], label='0')
patch5 = mpatches.Patch(color=colormap[4], label='$\sigma$')
patch6 = mpatches.Patch(color=colormap[5], label='$2\sigma$')
patch7 = mpatches.Patch(color=colormap[6], label='$3\sigma$')

plt.legend(handles=[patch1, patch2, patch3, patch4, patch5, patch6, patch7])

plt.title("Diverging Scale of Residual Error By Census Block Group -  Urban and Suburban")
plt.show()
```
